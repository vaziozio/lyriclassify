```python
file = open("lyrics.pkl",'rb')
lyrics = pickle.load(file)
```

```python
songs = {}

for artist in lyrics:
    songs[artist] = [song for song in lyrics[artist]]
```

```python
def remove_keys(row, column):
    if row[column] != 'no_lyrics':
        song = [v for k, v in row[column].items()]
        return song[0].replace('\n', ' ')
    return ''
```

```python
def resolve_text(df):

    for column in df.columns:
        df[column] = df.apply(lambda x: remove_keys(x, column), axis=1)
    
    df = df.apply(lambda row: '_'.join(row.values.astype(str)), axis=1)
    
    df = pd.DataFrame(df, columns=['text'])
    
    return df
```

```python
# ARRUMAR P BALANCEAR OS DADOS ENTRE OS ESTILOS
lyrics_df = pd.DataFrame.from_dict(songs, orient='index').fillna('no_lyrics').T[0:80].T

lyrics_df = resolve_text(lyrics_df)
```

```python
styles = []
for style in estilos:
    for artist in lyrics_df.index:
        if artist in estilos[style]:
            styles.append(style)

lyrics_df['styles'] = styles
```

```python
lyrics_df
```

```python
def strip_accents(s):
    return ''.join(c for c in unicodedata.normalize('NFD', s)
                  if unicodedata.category(c) != 'Mn')

lyrics_df['text_clean'] = lyrics_df['text'].apply(lambda x: x.translate(str.maketrans('', '', string.punctuation)))

lyrics_df['text_clean'] = lyrics_df['text_clean'].apply(lambda x: strip_accents(x).lower())

x_train, y_train = lyrics_df['text_clean'], lyrics_df['styles']

#train data
lyrics, tag_ = x_train.values , y_train.values
#test data
#lyrics_test, tag_test = x_test , y_test

stopwords = nltk.corpus.stopwords.words('portuguese')

# Instantiating the CountVectorizer method
count_vector = CountVectorizer(stop_words=stopwords)

# Fitting the training data and then return the matrix
training_data = count_vector.fit_transform(lyrics)

# Transform testing data and return the matrix.
testing_data = count_vector.transform(tag_)

#creating and fitting NB classifier to our data
clf = MultinomialNB()
clf.fit(training_data , tag_)
```

```python
musica = input('escreva um exemplo de letra:\n')

testing_data = count_vector.transform([musica])

print(clf.predict(testing_data), clf.predict_proba(testing_data))
```
